```python
import requests
from bs4 import BeautifulSoup
import urllib

# Function to download images
def download_images(url, class_name, download_path):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    # print(soup)

    # Find all elements with the specified class
    images = soup.find_all("img")
    print(images)

    # Download each image
    for i, img in enumerate(images):
        try:
            img_url = img['src']
            print(img_url)
            img_url = urllib.parse.urljoin(url, img_url)
            img_data = requests.get(img_url).content
        except:
            pass
        

        with open(f"{download_path}/image_{i + 1}.jpg", 'wb') as f:
            f.write(img_data)

# Example usage
url = 'https://www.ikea.com/sa/en/cat/sofas-fu003/'
class_name = 'image plp-product__image plp-product__image--alt'  # Replace with the actual class name
download_path = 'D:/Competition/Datathon-2023/Code/image'  # Replace with your desired download path

download_images(url, class_name, download_path)
```
